```
nums = [int(input("Número: ")) for _ in range(10)]

def eh_primo(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

primos = [x for x in nums if eh_primo(x)]
print(primos)
```

```
mat = [[int(input()) for _ in range(3)] for _ in range(3)]

identidade = True
for i in range(3):
    for j in range(3):
        if (i == j and mat[i][j] != 1) or (i != j and mat[i][j] != 0):
            identidade = False

print("É identidade?", identidade)
```

```
m = [[float(input()) for _ in range(3)] for _ in range(3)]

det = (
    m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1]) -
    m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0]) +
    m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0])
)

print("Determinante:", det)
```

```
prod = {
    "Arroz": [10,12,11],
    "Feijão":[7,9,8],
    "Macarrão":[4,6,5]
}

maior_var = None
nome = ""

for p, precos in prod.items():
    variacao = max(precos) - min(precos)
    if maior_var is None or variacao > maior_var:
        maior_var = variacao
        nome = p

print("Maior variação:", nome)
```
